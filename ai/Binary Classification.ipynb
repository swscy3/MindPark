```python
# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 필요한 라이브러리를 불러옵니다.
import pandas as pd
import numpy as np
```

```python
# 엑셀 파일 로드
data = pd.read_excel("processed_data (1).xlsx", header=0)

# pandas DataFrame을 numpy 배열로 변환
dataset = data.to_numpy()

print(data)
```

```python
# 데이터를 불러 옵니다.
# y 변수에 'Admission1' 열 저장
y = data['Admission1'].to_numpy()
# x 변수에 'Admission1'을 제외한 나머지 열 저장
x = data.drop(columns=['Admission1']).to_numpy()
```

```python
# x와 y 데이터 형태 확인
input_dim = x.shape[1]  # x의 열 개수
output_dim = 1  # y가 이진 분류(0 또는 1)일 경우

# 딥러닝 모델 설계
model = Sequential()
model.add(Dense(64, input_dim=input_dim, activation='relu'))  # 입력층 및 첫 번째 은닉층
model.add(Dense(32, activation='relu'))  # 두 번째 은닉층
model.add(Dense(output_dim, activation='sigmoid'))  # 출력층

# 모델 구조 출력
model.summary()
```

```python
# 학습을 위한 환경 설정
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# 테스트 데이터 설정 (x_test와 y_test는 분리된 테스트 데이터)
# 학습 데이터에서 일부를 테스트 데이터로 나눠야 합니다.
from sklearn.model_selection import train_test_split

# 데이터 분리 (훈련 데이터와 테스트 데이터로 분리)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 모델 재학습
model.fit(x_train, y_train, epochs=20, batch_size=10, verbose=1)

# 테스트 데이터 평가
loss, accuracy = model.evaluate(x_test, y_test)

# 테스트 정확도 출력
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
import pandas as pd
import numpy as np

# 테스트 데이터 예측
y_pred = model.predict(x_test)

# 예측값 이진화 (임계값 0.5)
y_pred_binary = (y_pred > 0.5).astype(int)

# 실제 테스트 데이터와 예측값 병합
results = pd.DataFrame({
    "BT": x_test[:, 0],  # 예시: 첫 번째 특성이 BT라고 가정
    "HR": x_test[:, 1],  # 예시: 두 번째 특성이 HR라고 가정
    "Real State": ["Abnormal" if val == 1 else "Normal" for val in y_test],
    "Predicted State": ["Abnormal" if val == 1 else "Normal" for val in y_pred_binary.ravel()]
})

# 정확도 계산
accuracy = np.mean(y_test == y_pred_binary.ravel()) * 100

# 정확도 추가
results.loc[len(results)] = ['-', '-', 'Accuracy (%)', f"{accuracy:.2f}%"]

# 결과 출력
print(results.to_string(index=False))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

# 데이터 정규화
scaler = StandardScaler()
x = scaler.fit_transform(x)

# 모델 생성 함수
def create_model():
    model = Sequential()
    model.add(Dense(64, input_dim=x.shape[1], activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# KerasClassifier로 모델 래핑
wrapped_model = KerasClassifier(model=create_model, epochs=20, batch_size=10, verbose=0)

# 교차 검증 실행
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 교차 검증 실행
results = cross_val_score(wrapped_model, x, y, cv=kfold)

# 결과 출력
print(f"Cross-validation Accuracy: {results.mean() * 100:.2f}% (+/- {results.std() * 100:.2f}%)")
```
